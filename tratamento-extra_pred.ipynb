# Bibliotecas

```python
import numpy as np
import pandas as pd
from IPython.display import display
```

# Importação dos Dados

```python
df = pd.read_csv('NPS_Complete_UFF_2022-07-18_2022-12-31_pre-tratado.csv', sep=';')

display(df)
```

# Tratamento Extra

### Transformação Logarítmica (p/ outliers)

```python
# Fazer uma transformação logarítmica para reduzir a influência dos outliers 
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col] = np.log(df[col] + 0.001) # adicionando 0.001 para evitar log(0)
```

### Transformação de colunas categóricas em binárias

```python
customer_segment = df['CUSTOMER_SEGMENT'] # coluna com dados de texto
df = df.drop(columns=['CUSTOMER_SEGMENT']) # exclui ela do dataframe
CUSTOMER_SEGMENT_encoded = pd.get_dummies(customer_segment, prefix='CUSTOMER_SEGMENT') # transforma em binárias
df = pd.concat([df, CUSTOMER_SEGMENT_encoded], axis=1) # adiciona no df 

# userq2_value = df['USERVALUEQ2_VALUE'] # coluna com dados de texto
# df = df.drop(columns=['USERVALUEQ2_VALUE']) # exclui ela do dataframe
# USERVALUEQ2_VALUE_encoded = pd.get_dummies(userq2_value, prefix='USERVALUEQ2_VALUE') # transforma em binárias
# df = pd.concat([df, USERVALUEQ2_VALUE_encoded], axis=1) # adiciona no df 

nps_class = df['NPS_Class'] # coluna com dados de texto
df = df.drop(columns=['NPS_Class']) # exclui ela do dataframe
NPS_Class_encoded = pd.get_dummies(nps_class, prefix='NPS_Class') # transforma em binárias
df = pd.concat([df, NPS_Class_encoded], axis=1) # adiciona no df 

survey = df['SURVEY'] # coluna com dados de texto
df = df.drop(columns=['SURVEY']) # exclui ela do dataframe
SURVEY_encoded = pd.get_dummies(survey, prefix='SURVEY') # transforma em binárias
df = pd.concat([df, SURVEY_encoded], axis=1) # adiciona no df 

term_fabr = df['term_fabr'] # coluna com dados de texto
df = df.drop(columns=['term_fabr']) # exclui ela do dataframe
term_fabr_encoded = pd.get_dummies(term_fabr, prefix='term_fabr') # transforma em binárias
df = pd.concat([df, term_fabr_encoded], axis=1) # adiciona no df 

device_tech = df['device_tech'] # coluna com dados de texto
df = df.drop(columns=['device_tech']) # exclui ela do dataframe
device_tech_encoded = pd.get_dummies(device_tech, prefix='device_tech') # transforma em binárias
df = pd.concat([df, device_tech_encoded], axis=1) # adiciona no df 

qtd_crm = df['Qtd_CRM'] # coluna com dados numéricos, mas que serão transformados em binários
df = df.drop(columns=['Qtd_CRM']) # exclui ela do dataframe
qtd_crm_encoded = qtd_crm.apply(lambda x: 1 if x > 0 else 0) # passa tudo que for maior que 0 para 1
df = pd.concat([df, qtd_crm_encoded], axis=1) # adiciona no df
```

# Exportar df Final
